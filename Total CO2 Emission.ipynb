```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_excel(r"C:\Users\sagar\OneDrive\Desktop\New folder (2)\Final Python Files\Final co2 emission file.xlsx")
df
```

```python
df.shape
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()

for i in df.select_dtypes(include='object'):

    df[i]=le.fit_transform(df[i])
df
```

```python
df.isna().sum()
```

```python
#saparate data into x(input or independent variable) and y(output or dependent variable)
x=df.drop(['total_co2_emission_including_luc'], axis=1)
y=df.total_co2_emission_including_luc
```

```python
# Split data into train and text
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.2,random_state=42)
```

```python
#build the model
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(x_train,y_train)
```

```python
# Make predictions on the test set
y_pred = model.predict(x_test)
```

```python
df2=pd.DataFrame({'y_test':y_test,'y_pred':y_pred})
df2
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
# Evaluate the model's performance
mse = mean_squared_error(y_test, y_pred)  # Mean Squared Error
rmse = mse**0.5  # Root Mean Squared Error
r2 = r2_score(y_test, y_pred)  # R-squared score
```

```python
# Print evaluation metrics
print(f"Mean Squared Error (MSE): {mse}")
print(f"Root Mean Squared Error (RMSE): {rmse}")
print(f"R-squared (R2): {r2}")
```

```python
# Create a DataFrame to display feature importance
importance_df = pd.DataFrame({
    'Feature': x.columns,
    'Importance': np.abs(model.coef_)
}).sort_values(by='Importance', ascending=False)

# Display feature importance
importance_df.head(59)
```

**Interpret Insights:**
- Top Drivers of Emissions:
Features with the highest coefficients (in absolute value) are the most significant drivers of emissions. For instance, Oil_co2, Coal_co2, and Land_use_change_co2 have the largest coefficients, indicating that they are key factors influencing emissions. These features contribute substantially to the overall emissions.

- Positive Coefficients:
A positive coefficient indicates a direct relationship: as the feature value increases, emissions also increase. In this case, all the features have positive coefficients, meaning that every feature contributes to higher emissions. This suggests that all these factors, including Oil_co2, Coal_co2, and Land_use_change_co2, play a role in driving emissions upward.

- Negative Coefficients:
A negative coefficient signifies an inverse relationship, meaning that as the feature increases, emissions decrease. While there are no negative coefficients in this analysis, a feature such as renewable_energy_use could potentially have a negative coefficient in a different scenario, suggesting that increasing renewable energy usage reduces emissions.

```python
# Sort the DataFrame by importance for better visualization
importance_df_sorted = importance_df.sort_values(by='Importance', ascending=False)

plt.figure(figsize=(10, 15))
sns.barplot(
    data=importance_df_sorted,
    x='Importance',
    y='Feature',
    palette='viridis'
)
plt.title("Feature Importance")
plt.xlabel("Importance")
plt.ylabel("Feature")
plt.tight_layout()
plt.show()
```
